```python
import pandas as pd 
```

```python
df = pd.DataFrame({"Size":["s","xxl","xl","l","s","s","m","xl","xl","l","l","l","s","m","s","m","s","s","xl","l","l","m","xl","l",]})
df
```

```python
df.head(4)
```

```python
ord_data = [["s","m","l","xl","xxl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)            # ordinal encoding through sklearn 
oe.fit(df[["Size"]])
```

```python
df["en_Size"] = oe.transform(df[["Size"]])
```

```python
df.head(10)
```

```python
# through map function 
```

```python
data = {"s":0,"m":1,"l":2,"xl":3,"xxl":4,}          # first create a dictionary 
```

```python
df["en_Size_map"] = df["Size"].map(data)            # column name -> . function -> parse the parameter 
```
