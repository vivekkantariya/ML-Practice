```python
import numpy as np 
```

```python
a = np.zeros((3,3))
b = np.random.rand(3,3)
c = np.ones((3,3))
d = np.array([1,2,3,4,5,16])
```

```python
a,b,c,d
```

```python
row = a[0:]                                                     # indexig   
row
```

```python
square = np.sqrt(d)                                            # Inbuilt Mathematical function 
square
```

```python
sum = np.sum(d)                                            # Inbuilt Mathematical function 
mean = np.mean(d)                                            # Inbuilt Mathematical function 
meadian = np.median(d)                                            # Inbuilt Mathematical function 
std = np.std(d)                                            # Inbuilt Mathematical function 
sum,mean, meadian, std
```

```python
dot = np.dot(c,b)
dot
```

```python
x = np.linspace(0,2**np.pi,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```
